```python
import PIL
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = PIL.Image.open('D:\ConorMcgregor.jpg')
txt =input('Enter message\n')
key = int(input('Enter Key'))
#ENC
nimg = np.array(img)
red = nimg[:,:,0]
lent = len(txt)
ascii_values = []
for char in txt :
    ascii_values.append(ord(char)+key)
entxt = np.array(ascii_values)
for i in range(len(entxt)):
    red[i,i] = entxt[i]
nimg[:,:,0]=red
#DECODE
decode=[]
red2=nimg[:,:,0]
for i in range(lent):
    decode.append(red2[i,i]-key)
ndeco = []
for x in decode:
    ndeco.append(x)
res = ""
for val in ndeco:
    res = res + chr(val)
#print ("Decoded string is : ", str(res))
opt = int(input('Enter 1 for Encode\nEnter 2 for Decode\n'))
if opt == 1:
    print('*******ENCODING********')
    print('Encoded Picture')
    plt.imshow(nimg)
elif opt == 2:
    print('*******DECODING********')
    print('Image to be Decrypted')
    plt.imshow(nimg)
    print('The Decoded message from the Steganographic image is')
    print(str(res))
else:
    print('Invalid input')
```
