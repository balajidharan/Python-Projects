### VISUAL REPRESENTATIONS

```python
group = forbes.groupby(['Nationality']).mean(['earnings ($ million)']).apply(list).reset_index()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(group,hue='Nationality',height=4).map(plt.scatter,'earnings ($ million)','Year').add_legend()
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(forbes,hue='Nationality',height=4).map(plt.scatter,'earnings ($ million)','Year').add_legend()
plt.show()
```

```python
plt.figure(figsize=(20,9))
plt.bar(list(forbes["Nationality"].value_counts().keys()),list(forbes["Nationality"].value_counts()))
plt.title("COUNTRY WISE PLAYERS")
plt.ylabel("count")
plt.xlabel("countries")
plt.show()
```
